# Steps to open the Apple store csv 
- initialize a variable called file then use `open()` with `utf8` encoding
- use csv `reader()` to create an iterable object 
- transform the file to a `list()` of lists
- test data set first 3 data rows and the header row
- OPEN-CSV.READER-LIST

```python
from csv import reader
opened_file = open('AppleStore.csv', encoding='utf8')
read_file = reader(opened_file)
data_set = list(read_file)

data_set[:4]
```
